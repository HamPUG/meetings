The decimal values 65 through 90 will display on a terminal the characters A through Z. Normally, for the English language, decimal values below 127 are all that is required for displaying characters on a terminal. The range between 127 and 255 are for European languages where certain characters have acutes, graves, circumflexes, etc.

There are sets of line drawing characters. They reside in the range from 9480 through 9628. For example the python code `print(chr(9492))` will display the character that is a bottom left corner of a rectangle:

`print(chr(9492))` displays └ 

The code below displays the characters in the low range from 32 to 255. It then displays all the characters in the   9001 to 10000 range. The line drawing sets may be identified within this higher range of outputs.

```python
s = ""
for i in range(32, 127):
    s += (chr(i))
print("Printable characters in the range 32 to 127:\n{}\n".format(s))

s = ""
for i in range(128, 255):
    s += (chr(i))
print("Printable characters in the range 128 to 255:\n{}\n".format(s))

# Display all characters in the range 9001 to 10000
count = 0
start = 9001
end = 10000
print("Printable characters in the range {} to {}:".format(start, end))   
s = str(start) + ": "
for i in range(start, end):
    count += 1
    if count == 50:
        s += chr(i)
        print("{}".format(s))
        s = "{} :".format(i)
        count = 0
    else:
        s += chr(i)
```

### Line Drawing Sets of Characters

There are primarily three sets of line drawing characters. A set drawn with a single line, a set with double lines, and a set with thick lines. Two complimenting sets of characters exists. One allows single line drawings to join to double line characters, and the other allows single line drawings to join with thick lines.

The following program displays the 5 sets of characters:

```python
# Single line drawing set:
single_list = [
    9472, 9474, 9484, 9488, 9492, 9496, 9500, 9508, 9516, 9524, 9532, ]
#    ─     │     ┌     ┐     └     ┘     ├     ┤     ┬     ┴     ┼ 

print("\nSingle total:", len(single_list))
s = ""
for i in single_list:
    s += chr(i) + " "
print(s)

# Double line drawing set:
double_list = [
    9552, 9553, 9556, 9559, 9562, 9565, 9568, 9571, 9574, 9577, 9580, ]
#    ═     ║     ╔     ╗     ╚     ╝     ╠     ╣     ╦     ╩     ╬  

print("\nDouble total:", len(double_list))
s = ""
for i in double_list:
    s += chr(i) + " "
print(s)

# Double - single converters drawing set
double_single_list = [
    9554, 9555, 9557, 9558, 9560, 9561, 9563, 9564, 9566, 9567, 9569, 9570, 9572, 9573, 9575, 9576, 9578, 9579, ]
#    ╒     ╓     ╕     ╖     ╘     ╙     ╛     ╜     ╞     ╟     ╡     ╢     ╤     ╥     ╧     ╨     ╪     ╫      

print("\nDouble to single converters total:", len(double_single_list))
s = ""
for i in double_single_list:
    s += chr(i) + " "
print(s)

# Solid lines
solid_list = [
    9473, 9475, 9487, 9491, 9495, 9499, 9507, 9515, 9523, 9531, 9547, ]
#    ━     ┃     ┏     ┓     ┗     ┛     ┣     ┫     ┳     ┻     ╋   

print("\nSolid lines total:", len(solid_list))
s = ""
for i in solid_list:
    s += chr(i) + " "
print(s)

# Solid to single converters
solid_single_list = [
    9485, 9486, 9489, 9490, 9493, 9494, 9497, 9498, 9501, 9502, 9503, 9504, 9505, 9506, 9509,
    9510, 9511, 9512, 9513, 9514, 9517, 9518, 9519, 9520, 9521, 9522, 9525, 9526, 9527, 9528, 
    9529, 9530, 9533, 9534, 9535, 9536, 9537, 9538, 9539, 9540, 9541, 9542, 9543, 9544, 9545, 9546, ]
#    ┍     ┎     ┑     ┒     ┕     ┖     ┙     ┚     ┝     ┞     ┟     ┠     ┡     ┢     ┥     
#
#    ┦     ┧     ┨     ┩     ┪     ┭     ┮     ┯     ┰     ┱     ┲     ┵     ┶     ┷     ┸     
#
#    ┹     ┺     ┽     ┾     ┿     ╀     ╁     ╂     ╃     ╄     ╅     ╆     ╇     ╈     ╉     ╊   

print("\nSolid to single converters total:", len(solid_single_list))
s = ""
for i in solid_single_list:
    s += chr(i) + " "
print(s)
```
