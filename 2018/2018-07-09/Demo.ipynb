## Getting data from HTTP

```python
#df = pd.read_html("https://en.wikipedia.org/wiki/Number_of_guns_per_capita_by_country")
df = pd.read_html("https://en.wikipedia.org/wiki/Number_of_guns_per_capita_by_country")[3]
```

```python
# Place holder to do stuff
```

```python
# placeholder
df
```

## Cleaning

```python
df.columns=['Country','PerResident','Dummy']
```

```python
df
```

```python
df = df[['Country','PerResident']]
```

```python
df
```

```python
df = df.iloc[1:]
```

```python
df
```

```python
df.PerResident = df.PerResident.str.split('[').str[0]
```

```python
df
```

```python
df['PerResident'] = df['PerResident'].astype('float64')
```

```python
# Place holder
df
```

## Save Cleaned Output

```python
df.to_csv('GunControl.csv')
```

# Plotting
Create a new Dataframe dfHigh of those greater than 20, sort it in ascending orderm and plot as a Barchart

```python

dfHigh = df[df.PerResident >= 20.0]

dfHigh = dfHigh.sort_values('PerResident')
```

```python
dfHigh
```

```python
import matplotlib
matplotlib.rc('xtick',labelsize=50)
matplotlib.rc('ytick',labelsize=50)

dfHigh.plot(kind='bar',x='Country',y='PerResident', legend = True,color='green',figsize=(90,30))
#dfHigh.plot(kind='pie',x='Country',y='PerResident',figsize=(60,60),labels=dfHigh['Country'],legend= False)
plt.show()  # GARY don't forget to uncomment this !!!!!
```
